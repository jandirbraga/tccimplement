# Análise Exploratória de Dados (EDA) 1: Diferentes modelos com todas as variáveis

**Objetivo da análise:** Identifificar relação entre os status de cada herói e definir como cada coluna será transformada no arquivo de transformação de dados.

**Arquivo analisado** ``hero_stats.json``

**Importing libraries**

```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
from matplotlib import pyplot as plt
```

**Loading file**

```python
data_path = '../data/raw_data'
game_stats_orig = pd.read_json(data_path+'/hero_stats.json')
game_stats = game_stats_orig.copy()
```

### Verificando tipos de dados e dados nulos

```python
game_stats.describe()
```

```python
game_stats.info()
```

**Análise de valores nulos**

```python
na_series = game_stats.isna().sum()
plt.subplots(figsize=(15,5))
plt.bar(x=na_series.index, height=na_series)
plt.xticks(rotation=90)
plt.title('Quantidade de valores nulos por variável')
plt.ylabel('Quantidade de amostras nulas')
plt.show()
```
