## Transformar colunas

Transformar as seguintes colunas:

- primary_attr     
- attack_type

### Variável ``primary_attr``
Transformar a coluna ``primary_attr`` em três colunas binárias:

1. primary_attr_int
2. primary_attr_str
3. primary_attr_agt

Depois deletar a coluna ``primary_attr`` 

```python
game_stats['primary_attr'].value_counts()
```

**Criando as novas colunas**

```python
aux_primary_df = pd.get_dummies(game_stats.primary_attr, prefix='primary_attr')
print(aux_primary_df.head())
```

```python
print(game_stats.shape)
game_stats = pd.concat([game_stats, aux_primary_df], axis=1)
print(game_stats.shape)
game_stats.head()
```

### Variável ``attack_type``
1. Transformar a coluna ``attack_type`` em ``is_Melee`` com 1 para Melee e 0 para o contrário.

```python
game_stats['attack_type'].value_counts()
```

```python
print(game_stats.shape)
game_stats.rename(columns={'attack_type':'is_Melle'}, inplace=True)
game_stats.is_Melle.replace({'Melee': 1, 'Ranged': 0}, inplace=True)
game_stats.head()
```

### Variável ``roles``
1. Descobrir a quantidade de roles únicos existentes
2. Dependendo da quantidade avaliar se é melhor usar one-hot encoding ou uma mudança cardinal

```python
game_stats['roles']
```

**Identificando quantidade de roles únicos**

Como são 9 roles, pode-se criar uma variável para cada um deles 

```python
s_roles = game_stats['roles'].copy()
single_list = list(itertools.chain.from_iterable(s_roles))
roles_set = set(single_list)

print('Distinct roles: ', roles_set)

distinct_roles = ['Nuker', 'Disabler', 'Initiator', 'Durable', 'Support', 'Jungler', 'Carry', 'Pusher', 'Escape']
```

```python
df_aux = pd.DataFrame(game_stats.roles.tolist(), index=game_stats.index).copy()
df_aux.head()
```

```python
df_one_hot_roles = pd.DataFrame(np.zeros([df_aux.shape[0], len(distinct_roles)]), columns=distinct_roles)

for i in np.arange(df_aux.shape[1]):
    aux_pd = pd.DataFrame()
    for j, role_name in enumerate(distinct_roles):
        df_one_hot_roles[role_name] += (df_aux.loc[:,i] == role_name).astype(int)

print('shape:', df_one_hot_roles.shape)
df_one_hot_roles.head()
```

**Conferindo se o valor do dataframe é equivalente à todos os roles da base**

```python
print(df_one_hot_roles.sum().sum(), len(single_list))
print(df_one_hot_roles.sum().sum() == len(single_list))
```

**Anexando ao dataframe**

```python
g = game_stats.copy()
```

```python
game_stats = g.copy()
game_stats = pd.concat([game_stats, df_one_hot_roles], axis=1)
game_stats
```
